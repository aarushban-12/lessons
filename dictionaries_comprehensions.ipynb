## Dictionaries are data structures that store key-value pairs

```python
my_dict = {
    "name": "Alice",
    "job": "Engineer",
    "age": 26
}

# Get values from the dictionary

print(my_dict["name"])
print(my_dict.get("age"))

# Update values in the dictionary
my_dict["city"] = "Charlotte"
my_dict["age"] = 27
print(my_dict)

# Remove values from the dictionary
job = my_dict.pop("job")
print(job, my_dict)
del my_dict["city"]
print(my_dict)
```

## Dictionary Methods

```python
print(my_dict.keys())

print(my_dict.values())

print(my_dict.items())

my_dict.update({"country": "USA", "hobby": "painting"})
print(my_dict)
print()

# Iterate through the dictionary
for key, value in my_dict.items():
    print(f"{key}: {value}")

# Dictionary Comprehension

doubles = {x: x*2 for x in range(10,30) if x % 5 == 0}
print(doubles)
print()

# Dictionaries can be nested inside other dictionaries

nested_dict = {
    "person1": {
        "name": "Alice",
        "age": 26
    },
    "person2": {
        "name": "Bob",
        "age": 30
    }
}

print(nested_dict["person2"]["age"])
```

## List Compherensions
This is a shorthand way to create lists through an iterable

```python
my_list = [x**2 for x in range(10) if x % 2 == 0]
print(my_list)

# Comprehensions can have if-else conditions

even_odd = [ ("even" if x % 2 == 0 else "odd") for x in range(10)]
print(even_odd)

# Comprehensions can also be nested
nested_comprehension = [ [x**2 for x in range(4)] for _ in range(3)]
for row in nested_comprehension:
    print(row)
```

Set comprehensions work the same way as list comprehensions

```python
my_set = {x for x in range(10) if x % 2 == 0}
print(my_set)

#Sets automatically remove duplicates
nums = [1, 2, 2, 3, 4, 4, 5]
unique_squares = {x**2 for x in nums}
print(unique_squares)

```

### Dictionaries can also be made using comprehensions

```python
keys = ["name", "age", "city"]
values = ["Alice", 26, "Charlotte"]
my_dict = {k: v for k, v in zip(keys, values)}
print(my_dict)

new_dict = {k: v for k, v in my_dict.items() if v != 26}
print(new_dict)
```

### Generator expressions can be used to iterate once while saving memory.

```python

gen = (x**2 for x in range(100) if x % 2 == 0)
print(gen)
for num in gen:
    print(num)
```
