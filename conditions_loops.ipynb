# Python Conditional Statements

## If-else Statements

The if-else statement allows for a code block to be executed based on a condition and a different code block to be executed otherwise

```python
number = 12

if number < 16:
    print("The number is less than 16.")
else:
    print("The number is greater than or equal to 16.")    
```

### If-elif-else statements allow for conditions to be chained

```python
my_list = [1, 2, 3, 4, 5]

#The conditions go down the line so each condition is checked 
# only if the previous one is false

if len(my_list) >= 6:
    print("Long list")
elif len(my_list) >= 4:
    print("Medium list")
else:
    print("Short list")    

#If statements can also be nested inside each other
age = 25
if age >= 18:
    if age < 21:
        print("You are an adult but cannot drink alcohol.")
    else:
        print("You are an adult and can drink alcohol.")
else:
    print("You are a minor.")        
```

## Logical Operators

| Operator | Description | Example |
|----------|-------------|----------|
| `and` | All must be True | `True and False` → `False` |
| `or` | At least one must be True | `True or False` → `True` |
| `not` | Negates the value | `not True` → `False` |

```python
my_num = 10
working = False

if not(my_num < 5 or working):
    print("Logical operators can be combined")
else:
    print("if statement was not true")
```

## Ternary Operators

Shorthand way of writing if statements

```python
score = 85

grade = "A" if score >= 90 else "F"
print(f"The grade is a {grade}.")
```

# Python Loops

## For Loops
The for loop iterates over the given iterable

```python
for character in "Hello":
    print(character)

# range() gives a sequence of numbers which can be used in a for loop

for i in range(1,7,2):
    print(i)

# enumerate() gives both the index and the value of each item in a list
my_list = ['apple', 'banana', 'cherry', 'date', 'elderberry']
for index, value in enumerate(my_list):
    print(f"Index: {index}, Value: {value}")

        
```

## While Loops
A while loop keeps on running while its condition is true

```python
count = 0
while count < 5:
    print(f"Count is {count}")
    count += 1
print()
# A break statement can be used to exit a loop early

for i in range(10):
    if i == 5:
        break
    print(i)
print()
# A continue statement can be used to skip the current iteration
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

## Nested Loops

```python
for i in range(5):
    for j in range(5-i):
        print("*", end="")
    print()

print()
# Loops can also have multiple conditions in their headers

my_list = ["apple", "banana", "cherry", "date", "elderberry"]
for i in range(len(my_list)):
    if i % 2 == 0 and len(my_list[i]) > 5:
        print(my_list[i])
```
