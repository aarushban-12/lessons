## Variable Naming Conventions (PEP 8)

**Valid Variable Names:**
- Must start with a letter (a-z, A-Z) or underscore (_)
- Can contain letters, numbers, and underscores
- Case-sensitive (`age`, `Age`, `AGE` are different)

**Naming Styles:**
- `snake_case` - Recommended for variables and functions
- `UPPER_CASE` - Constants
- `CamelCase` - Classes

**Rules:**
- Cannot start with a number
- Cannot contain spaces or special characters (except _)
- Avoid Python keywords (if, for, while, etc.)

```python
# Variable names

hello = 125
_1235 = "python"

print(hello, _1235)

# Invalid variable names
# 123hello = 125
# .name = "python"
```

```python
#Variable data types and conversion

var_int = 10
var_float = 3.14
var_str = "Hello"
var_bool = True

print(var_int, var_float, var_str, var_bool)

#Data types can be converted using casting

print(type(var_int))

var_int = float(var_int)

print(var_int, type(var_int))

x = 250
y = 5.25
quotient = x / y
print(quotient, type(quotient))
```

```python
#Use del keyword to delete variables

my_var = "This variable will be deleted"
print(my_var)

del my_var

try:
    print(my_var)
except NameError:
    print("Error will be triggered since variable has been deleted")    
```

```python
#Shorthand assignment operators

my_int = 100

my_int += 50
print(my_int)

my_int -= 25   
print(my_int)

#This works with other data types as well

my_str = "Hello"
my_str += " Everyone"
print(my_str)
```

## Python Strings

Python strings are immutable and can be created using single, double, or triple quotes. 

```python
string1 = "Hello"

string2 = 'World'

string3 = """This is a string that can 
span multiple lines."""

print(string1)
print(string2)
print(string3)
```

```python
#Strings are indexed and can be accessed using square brackets 
# indexing starts at 0

first_char = string1[0]
last_char = string1[-1]

print(first_char, last_char)

# String Slicing - [start:end:step]

words = "this is some filler text"

print(words[0:4]) 
print(words[::2])
```

## String Methods

```python
# Strings can be concatenated using the + operator

text2 = "This is another string."
concatenated = words + " " + text2
print(concatenated)

# Replace words 
original = "The quick brown fox jumps over the lazy dog."
print(original.replace("fox", "dog"))

#Split into list
words = original.split()
print(words)

#Join list to string
print(" ".join(words))
```

```python
# Find index of first occurence of substring
text = "Hello world, welcome to the world!"
print(text.find("world"))

#Count occurences of substring
print(text.count("world"))

# Check if substring exists in string
print("welcome" in text)
print("python" in text)
```

## Use f-strings to format strings

```python
print(f"Aarush says {text}")

price = 12.18290
print(f"The price is {price:.2f}")
```

## Lists
Lists are collections of elements that are mutable, ordered, and allow duplicates

They are indexed and sliced the same way as strings

```python
my_list = [1, True, 3, "hello", 5.75]
print(my_list)

print(f"third element: {my_list[2]}")
print(f"last 3 elements: {my_list[2:]}")
```

```python
# Since lists are mutable, we can change their elements
my_list[1] = False
print(my_list)

my_list[0:2] = ["one", "two", "three", "four"]
print(my_list)
```

## Adding to lists

```python
my_list = [1,2]

my_list.append(3)
print(my_list)

my_list.insert(1, 1.5)
print(my_list)

my_list.extend([4,5])
print(my_list)

my_list += [6,7]
print(my_list)
```

## Removing from lists

```python
my_list.remove(1.5)
print(my_list)

my_list.pop()
print(my_list)

x = my_list.pop(0)
print(x, my_list)

del my_list[0:3]
print(my_list)

my_list.clear()
print(my_list)
```

## Other list methods

```python
my_list = [1,2,3,4,2,3,1,4,1]

# Count number of occurences of an element
print(my_list.count(1))

#Sort list
my_list.sort()
print(my_list)

#Reverse list
my_list.reverse()
print(my_list)

#Create copy of list
new_list = my_list.copy()
new_list.append(5)
print(my_list)
print(new_list)
```

## List comprehensions allow for a easy way to create lists using a condition applied to all elements

```python
my_list = [x for x in range(5)]
print(my_list)
```

## Tuples are similar to lists but they are immutable.

Elements can be accessed the same way as lists but cannot be modified

```python
nums = (1,2,3,4)
print(nums)

nums = nums + (5,)
print(nums)

# count number of occurences of an element in a tuple
nums = (1,2,3,1,4,14)
print(nums.count(1))

# get index of a value in a tuple
print(nums.index(4)) 
```

## Sets are unordered collections of elements that are mutable and only have unique elements

Sets do not use indexing and duplicates are automatically removed

```python
nums = {1,2,3,4}
print(nums)

nums.add(5)
print(nums)

nums.update([6,7,8])
print(nums)

nums.remove(2)
print(nums)

nums.discard(10) # does not raise error if element does not exist
print(nums)

x = nums.pop() # removes and returns an ramdom element
print(x, nums)


```

```python
# Set operations

set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

# Union - all elements from both sets
print(f"Union (|): {set_a | set_b}")
print(f"Union method: {set_a.union(set_b)}")

# Intersection - elements common to both sets
print(f"Intersection (&): {set_a & set_b}")
print(f"Intersection method: {set_a.intersection(set_b)}")

# Difference - elements in first set but not in second
print(f"Difference (-): {set_a - set_b}")
print(f"Difference method: {set_a.difference(set_b)}")
print(f"Reverse difference: {set_b - set_a}")

# Symmetric difference - elements in either set but not in both
print(f"Symmetric difference (^): {set_a ^ set_b}")
print(f"Symmetric difference method: {set_a.symmetric_difference(set_b)}")

# Subset and superset
set_c = {1, 2, 3}
print(f"Is {set_c} subset of {set_a}? {set_c.issubset(set_a)}")
print(f"Is {set_a} superset of {set_c}? {set_a.issuperset(set_c)}")

# Disjoint - no common elements
set_d = {10, 11, 12}
print(f"Are {set_a} and {set_d} disjoint? {set_a.isdisjoint(set_d)}")
```
